### MoNuSeg  SOTA Dice Coeff

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import random

x = [
    "Graham et al. (2019)",
    "Sahasrabudhe et al. (2020)",
    "Vahadane et al. (2021)",
    "Boserup and Selvan (2022)",
    "Ali et al. (2022)",
    "Luo et al. (2022)",
    "An et al. (2022)",
    "Dogar et al. (2023)",
]

dices = np.array(
    [
        0.8260,
        0.7477,
        0.8100,
        0.6797,
        0.8470,
        0.8025,
        0.8080,
        0.8150,
    ]
)
cmap = ["#{:06x}".format(random.randrange(256)) for i in range(8)]
plt.bar(x, dices, color=cmap)
plt.xticks(rotation=45)
plt.ylabel("Average Dice Coefficient (ADC)")
plt.title("MoNuSeg")
plt.show()
plt.savefig("monuseg_sota_dice.png")
```
